```python
#Importando pandas
import pandas as pd
```

```python
#Leitura dos arquivos
df1 = pd.read_excel("Aracaju.xlsx",engine="openpyxl")
df2 = pd.read_excel("Fortaleza.xlsx")
df3 = pd.read_excel("Natal.xlsx")
df4 = pd.read_excel("Recife.xlsx")
df5 = pd.read_excel("Salvador.xlsx")

```

```python
df5.head()
```

```python
#Juntado os arquivos
df = pd.concat([df1,df2,df3,df4,df5])
```

```python
#Exibindo as 5 ultimas linhas
df.tail()
```

```python
#Pegar amostra
df.sample(5)
```

```python
#Verificando o tipo de dado de cada coluna
df.dtypes
```

```python
#Alterando o tipo da coluna LojaId
df["LojaID"] = df["LojaID"].astype("object")
```

```python
df.head()
```

```python
#Consultando linhas com valores faltantes
df.isnull().sum()
```

```python
#Substintuindo os valores nulos pela média
df["Vendas"].fillna(df["Vendas"].mean(), inplace =True)
```

```python
df["Vendas"]
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.sample(15)
```

```python
#Criando acoluna de receita
df["Receita"] = df["Vendas"].mul(df["Qtde"])
```

```python
df["Receita/Vendas"] = df["Receita"]/ df["Vendas"]
```

```python
df.head()
```

```python
#Retorna a menor receita
df["Receita"].min()
```

```python
#Retorna a maior receita
df["Receita"].max()
```

```python
#nlargest top 3 com base em receita
df.nlargest(3, "Receita")
```

```python
#nlargest top 3 piores  receita
df.nsmallest(3, "Receita")
```

```python
#Agrupamento por cidade A loja que não esta indo muito bem
df.groupby("Cidade")["Receita"].sum()
```
